```python
from google.colab import files

# Upload two .wav files
print("Upload two .wav files of your voice for cloning:")
uploaded = files.upload()

# Extract file paths
reference_audio_1 = list(uploaded.keys())[0]  # First uploaded file
reference_audio_2 = list(uploaded.keys())[1]  # Second uploaded file

print(f"Uploaded files: {reference_audio_1}, {reference_audio_2}")
```

```python

from TTS.api import TTS

# Load the YourTTS model
print("Loading YourTTS model...")
tts = TTS(model_name="tts_models/multilingual/multi-dataset/your_tts", progress_bar=True)

```

```python
# Text to synthesize
text_to_synthesize = "Hello this is my cloned voice in english but i used dutch audio samples!"  # Replace with your text

# Output files
output_file_1 = "output_reference_1.wav"
output_file_2 = "output_reference_2.wav"

# Generate speech with the first reference
print("Generating speech with the first reference...")
tts.tts_to_file(
    text=text_to_synthesize,
    file_path=output_file_1,
    speaker_wav=reference_audio_1,
    language="en"  # Replace "en" with "nl" if supported
)

# Generate speech with the second reference
print("Generating speech with the second reference...")
tts.tts_to_file(
    text=text_to_synthesize,
    file_path=output_file_2,
    speaker_wav=reference_audio_2,
    language="en"  # Replace "en" with "nl" if supported
)
```
